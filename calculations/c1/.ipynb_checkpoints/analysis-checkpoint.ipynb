```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_ttmult_spectrum(fn):
    with open(fn) as fp:
        x = list()
        y = list()
        for line in fp:
            if 'Sticks' in line:
                break
            tokens = line.split()
            if tokens:
                x.append(float(tokens[0]))
                y.append(float(tokens[1]))

        next(fp)

        xsticks = list()
        ysticks = list()
        for line in fp:
            tokens = line.split()
            xsticks.append(float(tokens[0]))
            ysticks.append(float(tokens[1]))

    x = np.array(x)
    y = np.array(y)
    xsticks = np.array(xsticks)
    ysticks = np.array(ysticks)
    
    return x, y, xsticks, ysticks

def read_quanty_spectrum(fn, column=2):
    data = np.loadtxt(fn, skiprows=5)
    x = data[:, 0]
    y = data[:, column]
    return x, y
```

```python
fig, ax = plt.subplots(1, 1)

x, yz = read_quanty_spectrum('Quanty/input.spec', column=6)
_, yr = read_quanty_spectrum('Quanty/input.spec', column=8)
_, yl = read_quanty_spectrum('Quanty/input.spec', column=10)

yiso = yr + yl + yz
ax.plot(x, yiso, label='Quanty')

x, y, _, _ = read_ttmult_spectrum('TTMult/input_iso.xy')
ax.plot(x, y, label='TTMult')
ax.plot(x, y * 2, label='TTMult x 2')

ax.legend()
title = ax.set_title('XAS')
```

```python
fig, ax = plt.subplots(1, 1)

x, yr = read_quanty_spectrum('Quanty/input.spec', column=8)
_, yl = read_quanty_spectrum('Quanty/input.spec', column=10)
yxmcd = yr - yl
ax.plot(x, yxmcd, label='Quanty')

x, y, _, _ = read_ttmult_spectrum('TTMult/input_xmcd.xy')
ax.plot(x, y, label='TTMult')
ax.plot(x, y * 2, label='TTMult x 2')

ax.legend()
title = ax.set_title('XMCD')
```

```python
l = 2
n = 10 - 8.1022
# As defined in Maurits' thesis on page 58 and 65.
Lz_sum_rules = np.trapz(yxmcd, x) / np.trapz(yiso, x) * l * n
Lz_calculation = -0.1685
```

```python
print('Lz (calculation) = {:f}'.format(Lz_calculation))
print('Lz (sum rules)   = {:f}'.format(Lz_sum_rules))
```
