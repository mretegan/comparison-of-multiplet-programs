```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
import sys
import math

sys.path.insert(0, '../../utils')
from read_spectrum import read_ttmult_spectrum, read_quanty_spectrum
```

```python
SHIFT = 7121.216 # 7121.436
```

# Spectra

```python
fig, ax = plt.subplots()
ax.grid()

x, y = read_quanty_spectrum('Quanty/input_iso.spec')
ax.plot(x + SHIFT, y, label='Quanty')
xiso_quanty = x
yiso_quanty = y

x, y, xsticks, ysticks = read_ttmult_spectrum('TTMult/input-15iso300K.xy')
# The isotropic spectrum from TTMult must be devided by 15.
y = y / 15
ax.plot(x, y, label='TTMult')
xiso_ttmult = x
yiso_ttmult = y

plt.title('Isotropic')
ax.legend()
```

# Sum rule

15$\int_{Iso} = \frac{4l + 2 - n_d}{(2l + 1)}$ 

```python
nd_calculation = 6
nd_sum_rule = 10 - 15 * 5 * np.trapz(yiso_quanty, xiso_quanty)

print('nd (calculation) = {:f}'.format(nd_calculation))
print('nd (sum rule)    = {:f}'.format(nd_sum_rule))
```
