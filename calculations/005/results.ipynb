```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
import sys
import math

sys.path.insert(0, '../../utils')
from read_spectrum import read_ttmult_spectrum, read_quanty_spectrum
```

```python
SHIFT = -13.831 
```

# Spectra

```python
fig, ax = plt.subplots()
ax.grid()

x, y = read_quanty_spectrum('Quanty/input_iso_dip.spec')
ax.plot(x + SHIFT, y, label='Quanty')

x, y, xsticks, ysticks = read_ttmult_spectrum('TTMult/dipolar/input-3iso1K.xy')
y = y / 3
ax.plot(x, y, label='TTMult')

plt.title(r'Isotropic (1s $\rightarrow$ 4p)')
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.grid()

x, y = read_quanty_spectrum('Quanty/input_iso_quad.spec')
ax.plot(x + SHIFT, y, label='Quanty')

x, y, xsticks, ysticks = read_ttmult_spectrum('TTMult/quadrupolar/input-15iso1K.xy')
y = y / 15
ax.plot(x, y, label='TTMult')

plt.title(r'Isotropic (1s $\rightarrow$ 3d)')
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.grid()

x, y = read_quanty_spectrum('Quanty/input_iso_quad_scaled.spec')
ax.plot(x + SHIFT, y, label='Quadrupolar')

x, y = read_quanty_spectrum('Quanty/input_iso_dip_scaled.spec')
ax.plot(x + SHIFT, y, label='Dipolar')

x, y = read_quanty_spectrum('Quanty/input_iso.spec')
ax.plot(x + SHIFT, y, label='Total')

ax.set_xlim(-15, -10)
ax.set_ylim(0, 0.15)
ax.legend()

plt.title(r'Isotropic')
```
